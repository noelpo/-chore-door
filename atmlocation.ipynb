```python
import pandas as pd
import numpy as np
import haversine as hs 
import folium
```

```python
Atm_user = pd.read_excel('/Kingston_location.xlsx')
atm_location = pd.read_excel('/ATM_location.xlsx')
```

```python
Atm_user['coordinates']= list(zip(Atm_user.Lat,Atm_user.Lon))
atm_location['coordinates']= list (zip(atm_location.Lat, atm_location.Lon))
```

```python
atm_location.tail()
```

```python
def location_from(user,atm):
  dist=hs.haversine(user,atm)
  return round(dist,2)
```

```python
for _,row in atm_location.iterrows():
  Atm_user[row.ATM]= Atm_user['coordinates'].apply(lambda x: location_from(row.coordinates,x))
```

```python
Atm_user.head()
```

```python
def create_map(user,atm):
    m = folium.Map(location=[user.Lat.mean(), user.Lon.mean()], zoom_start=12, tiles='OpenStreetMap')
    
   
    
    for _, row in user.iterrows():
         
        if row['Condition']=='Out':
            cluster_colour='blue'
        else:
            cluster_colour = 'green'
        
        folium.CircleMarker(
            location= [row['Lat'],row['Lon']],
            radius=5,
            popup= row[['Condition','Location']],
            color=cluster_colour,
            fill=True,
            fill_color=cluster_colour
            ).add_to(m)
    
    for _, row in atm.iterrows(): 
        folium.Marker(
            location= [row['Lat'],row['Lon']],
            radius=5,
            popup= row['Atm'],
            color='red',
            fill=True,
            fill_color='red'
             ).add_to(m)
    for _, row in atm.iterrows(): 
        folium.Circle(
            location= [row['Lat'],row['Lon']],
            radius=3000,
            popup= row['Atm'],
            color='black',
            fill=False,
        ).add_to(m)
        
    return m
```

```python
ATM_list=['A','B','C','D','E','F','G','H']
```

```python
Atm_user['Condition']='Out'

for i in ATM_list:
    Atm_user['Condition'][Atm_user.Condition=='Out']=Atm_user[Atm_user.Condition=='Out'][i].apply(lambda x: 'In' if x<3 else 'Out')
    
```

```python
m= create_map(Atm_user,atm_location)
m
```
